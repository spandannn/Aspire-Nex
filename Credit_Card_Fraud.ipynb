# Checking how many transaction is legit and fraud
* Class 0: Legit
* Class 1: Fraud

```python
data['Class'].value_counts()
```

```python
legit = data[data.Class == 0]
fraud = data[data.Class == 1]
```

```python
legit.shape
```

```python
fraud.shape
```

```python
data.groupby('Class').mean()
```

# Generating same number of samples in legit data to balace the dataset

```python
legit_sample = legit.sample(n=492)
legit_sample
```

# Merging both legit and fraud datasets in a single dataframe using pd.concat()

```python
data_merged = pd.concat([legit_sample, fraud], axis = 0)
data_merged.shape
```

# Checking if data is balanced or not

```python
data_merged['Class'].value_counts()
```

```python
data_merged['Class'].describe()
```

# Splitting the data into features and target

```python
x = data_merged.drop(['Class'], axis = 1)
y = data_merged['Class']
```

#  Split the data into training and testing sets
 (80% training dataset and 20% testing dataset)

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size = 0.8, random_state = 42)
```

```python
algorithms = [
    ('Logistic Regression', LogisticRegression(max_iter=1000)),
    ('Decision Tree', DecisionTreeClassifier()),
    ('Random Forest', RandomForestClassifier())
]
```

# Evaluating different models and printing the results

```python
for name, algorithm in algorithms:
  algorithm.fit(x_train, y_train)
  y_pred = algorithm.predict(x_test)
  accuracy = accuracy_score(y_test, y_pred)
  report = classification_report(y_test, y_pred)
  matrix = confusion_matrix(y_test, y_pred)
  print(f'\n{name} Results:')
  print(f'Accuracy: {accuracy*100}%')
  print(f'Classification Report:\n{report}')
  print(f'Confusion Matrix:\n{matrix}')
  print('='*55)
```
